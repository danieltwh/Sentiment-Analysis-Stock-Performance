```python
import requests

# from config import ALPHA_VANTAGE_API_KEY
from mlconfig import ALPHA_VANTAGE_API_KEY
```

```python
STOCK_TICKER = "AAPL"

# In YYYYMMDDTHHMM format
DATE_START = "20220401T0800"
DATE_END = "20220501T0800"
```

```python
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
# url = f'''https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={STOCK_TICKER}&time_from={DATE_START}&time_end={DATE_END}&apikey={ALPHA_VANTAGE_API_KEY}'''
```

```python
url = "https://www.alphavantage.co/query"
```

```python
payload = {
    "function": "NEWS_SENTIMENT",
    "tickers": STOCK_TICKER,
    "time_from" : DATE_START,
    "time_to" : DATE_END,
    "limit": 200,
    "apikey" : ALPHA_VANTAGE_API_KEY
}
```

```python
resp = requests.get(url, params=payload)
```

```python
data = resp.json()

# print(data)
```
